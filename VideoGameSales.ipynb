```python
# import pandas

import pandas as pd
```

```python
# reading the csv file using read_csv in pandas

df_videogame_sales = pd.read_csv( 'vgsales.csv' )

df_videogame_sales
```

```python
# retrieve the first 5 rows from the dataframe using the head() method

# by default the head() method returns the first 5 rows of the dataframe

df_videogame_sales.head( )
```

```python
# retrieve the last 5 rows from the dataframe using the tail() method

# by default the tail() method returns the first 5 rows of the dataframe

df_videogame_sales.tail( )
```

```python
# retrieve any number of rows from the dataframe using either the head( ) or tail() method 

df_videogame_sales.head( 7 )

# returns all the rows exclusive of the index 7
```

```python
# selecting a column from a dataframe 

df_videogame_sales[ 'Name' ]
```

```python
# selecting multiple columns from a dataframe 

df_videogame_sales[ [ 'Name', 'Year' ] ]
```

```python
# specifying conditionals within a column name

df_videogame_sales[ ( df_videogame_sales[ 'NA_Sales' ] > 20 ) ]
```

```python
# finding the datatype of all the columns in the dataframe using the dtypes() method

df_videogame_sales.dtypes
```

```python
# finding the datatype of a column in the dataframe using the dtype() function

df_videogame_sales['Year'].dtype
```

```python
# finding the highest index in the dataframe which is 1 less than the length of the dataframe using the max() method

max( df_videogame_sales.index )
```

```python
# finding the lowest index in the dataframe using the min() method

min( df_videogame_sales.index )
```

```python
# finding the dimensions of the dataframe using the shape() method

# left value is the number of rows

# right value is the number of columns

df_videogame_sales.shape
```

```python
# displaying all the rows in the dataframe using the set_option() method

pd.set_option( 'display.max_rows', None )
```

```python
# getting all the columns in the dataframe using the columns() method

df_videogame_sales.columns
```

```python
# sorting the dataframe column values according to ascending order order using the sort_values() method

# no need to specify ascending = True unless you want to explicit, the cell would run normally

df_videogame_sales['Genre'].sort_values( ascending = True )
```

```python
# sorting the dataframe column values according to descending order using the sort_values() method

df_videogame_sales['Platform'].sort_values( ascending = False )
```

```python
# Visualizing the data in the dataframe using the plot() method
 
df_videogame_sales.plot( kind='line', y=[ 'NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales', 'Global_Sales' ] )
```

```python
# creating a new column in a dataframe

# axis 1 is the y axis

# axis 0 is the x axis

df_videogame_sales[ 'Total_Sales' ] = df_videogame_sales[ 'EU_Sales' ] + df_videogame_sales[ 'NA_Sales' ] + df_videogame_sales[ 'JP_Sales' ]
```

```python
# dropping a column in the dataframe using the drop() method

df_videogame_sales.drop( [ 'Total Sales' ] )
```

```python
# getting some basic statistics on the dataframe using the describe() method

df_videogame_sales.describe()
```

```python
# rounding dataframe column values using round() method

round( df_videogame_sales.describe() )
```

```python
# finding the count of values in a dataframe column using the count() method

df_videogame_sales['Genre'].count()
```

```python
# finding the mean of a dataframe using the mean() method

df_videogame_sales.mean()
```

```python
# finding the median of a dataframe using the median() method

df_videogame_sales.median()
```

```python
# finding the mode of a dataframe using the mode() method

df_videogame_sales.mode(numeric_only=True)
```

```python
# filtering a dataframe using the filter() method

df_videogame_sales.filter(['Genre', 'Other_Sales'])
```

```python
# Access a dataframe column by label using loc() method

df_videogame_sales.loc[0]
```
